Contexto y Objetivos

# <a id="exploratory-analysis"></a> Análisis Univariado

```python
# Generar gráficas para analizar la distribución
for col in df[target].columns:
    if col in df[target].columns:
        fig, axes = plt.subplots(1, 3, figsize=(14, 5))

        # Histograma con KDE (curva de densidad)
        sns.histplot(df[col], bins=30, kde=True, ax=axes[0], color="royalblue")
        axes[0].set_title(f"Distribución de {col}")
        axes[0].set_xlabel(col)
        axes[0].set_ylabel("Frecuencia")

        # Boxplot para detección de outliers
        sns.boxplot(x=df[col], ax=axes[1], color="orange")
        axes[1].set_title(f"Boxplot de {col}")


        # Scatter para visualizar la dispersión de los datos
        plt.scatter(range(len(df[col])), df[col], color="green")
        axes[2].set_title(f"Dispersión de {col}")

        plt.tight_layout()
        plt.show()
```

```python

plt.figure(figsize=(15, 12))  # Ajusto el tamaño de la figura para que se vean las gráficas de forma cómoda

# Subgráfico 1: Distribución de 'Cultivo' (Barras horizontales)
plt.subplot(2, 2, 1)  # (filas, columnas, índice de la gráfica)
sns.countplot(y='Cultivo', data=df, color="royalblue") 
plt.title('Distribución de Cultivo')
plt.xticks(rotation=90)

# Subgráfico 2: Distribución de 'Producto' (Barras horizontales)
plt.subplot(2, 2, 2)
sns.countplot(y='Producto', data=df, color="royalblue")
plt.title('Distribución de Producto')
plt.xticks(rotation=90)

# Subgráfico 3: Distribución de 'Tipo Quema' (Barras horizontales)
plt.subplot(2, 2, 3)
sns.countplot(y='Tipo Quema', data=df, color="royalblue")
plt.title('Distribución de Tipo Quema')
plt.xticks(rotation=90)

# Subgráfico 4: Distribución de 'T.Corte' (Barras horizontales)
plt.subplot(2, 2, 4)
sns.countplot(y='T.Corte', data=df, color="royalblue")
plt.title('Distribución de T.Corte')
plt.xticks(rotation=90)

plt.tight_layout()
plt.show()
```

```python
# Calcular la matriz de correlación

var_numericas = df.select_dtypes(include=['float64', 'int64'])

# Validar la correlación en respuesto a la variable objetivo "TCH"
correlacion = var_numericas.corr()['TCH']
correlacion = correlacion.sort_values(ascending=False)

# Crear grafico
plt.figure(figsize=(8, 6))
sns.barplot(x=correlacion.values, y=correlacion.index, palette='coolwarm')

plt.title(f'Correlación de mi_variable con otras variables', fontsize=14)
plt.xlabel('Correlación', fontsize=10)
plt.ylabel('Variables', fontsize=10)
plt.xticks(fontsize=9)  
plt.yticks(fontsize=7)  
plt.show()
```
