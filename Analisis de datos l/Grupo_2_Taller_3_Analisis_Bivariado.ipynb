# <a id="contexto-title"></a> 2. Contexto


**Análisis de Factores de Riesgo Cardiovascular en el Distrito de Aguablanca, Cali: Una Evaluación Contextualizada del Framingham Score​:**

El riesgo cardiovascular es una preocupación de salud pública a nivel mundial. Herramientas como el Framingham Score han sido ampliamente utilizadas para estimar el riesgo de eventos cardiovasculares en diversas poblaciones. Sin embargo, su aplicabilidad en comunidades con características socioeconómicas y demográficas específicas, como el Distrito de Aguablanca en Cali, Colombia, requiere evaluación. Este proyecto tiene como objetivo identificar y analizar los factores de riesgo cardiovascular más significativos en esta población, comparándolos con los contemplados en el Framingham Score. **_Se realizará un análisis de variables como masa corporal, edad, género, enfermedad renal crónica y tensión arterial, considerando el contexto socioeconómico y cultural de la comunidad. Los resultados permitirán adaptar o desarrollar herramientas de evaluación de riesgo cardiovascular más precisas para poblaciones similares_**.

---

## Revisión Bibliográfica

A continuación, se presentan tres artículos científicos que abordan la evaluación del riesgo cardiovascular y la aplicabilidad del Framingham Score en diferentes contextos:​

**1. Evaluación del Riesgo de Enfermedad Cardiovascular: Perspectivas desde Framingham**

Este artículo destaca que las funciones de riesgo de Framingham combinan información de factores como sexo, edad, presión arterial sistólica, colesterol total, colesterol HDL, tabaquismo y diabetes para estimar el riesgo de desarrollar enfermedades cardiovasculares en un período determinado, como los próximos 10 años.

**2. Estado Socioeconómico, Factores de Riesgo Cardíaco y Enfermedad Cardiovascular**

Este estudio revela que la hipertensión es el factor de riesgo más común en grupos de bajo nivel socioeconómico. Además, factores como antecedentes familiares de enfermedades cardiovasculares, tabaquismo e hiperlipidemia también son prevalentes en distintos grupos socioeconómicos.

**3. Asociación del Ingreso Familiar con el Riesgo de Enfermedad Cardiovascular en Adultos Estadounidenses**

Este artículo analiza cómo el bajo ingreso familiar, una medida del estado socioeconómico, puede estar asociado con un mayor riesgo de enfermedades cardiovasculares, destacando la importancia de considerar factores socioeconómicos en la evaluación del riesgo.

**4. Sobre la presión arterial y su relación para determinar riesgo caridovascular**

Este artículo pone en contexto por qué medir la presión diferencial puede ayudar a un equipo de atención médica a predecir el riesgo de padecer eventos cardíacos y vasculares, como ataques cardíacos o accidentes cerebrovasculares. Una presión diferencial superior a 60 se considera factor de riesgo para enfermedad cardíaca, especialmente en personas mayores.

# <a id="univariate-analysis"></a> Análisis Univariado


Para cada variable clave:

1. Calcularemos estadísticas descriptivas (media, mediana, desviación estándar,
   valores mínimos y máximos).
2. Graficaremos su distribución con histogramas y boxplots para detectar posibles problemas.


```python
# Seleccion de variables
features = ['EDAD (Años cumplidos)','25_TenArtSis','26_TenArtDitlica', '23_Peso', '24_Talla']

# Verificar registros con caracteres especiales dentro del formato numerico
df[features] = df[features].apply(pd.to_numeric, errors='coerce')
df_clean = df[features] .dropna(subset=features)
df_clean.describe().T
```

```python

# Generar graficas para analizar la distribución
for col in df_clean.columns:
    if col in df_clean.columns:
        fig, axes = plt.subplots(1, 3, figsize=(14, 5))

        # Histograma con KDE (curva de densidad)
        sns.histplot(df_clean[col].dropna(), bins=30, kde=True, ax=axes[0], color="royalblue")
        axes[0].set_title(f"Distribución de {col}")
        axes[0].set_xlabel(col)
        axes[0].set_ylabel("Frecuencia")

        # Boxplot para detección de outliers
        sns.boxplot(x=df_clean[col], ax=axes[1], color="orange")
        axes[1].set_title(f"Boxplot de {col}")


        # Violinplot para visualizar la densidad de los datos
        sns.violinplot(x=df_clean[col], ax=axes[2], color="green")
        axes[2].set_title(f"Violinplot de {col}")

        plt.tight_layout()
        plt.show()
```
