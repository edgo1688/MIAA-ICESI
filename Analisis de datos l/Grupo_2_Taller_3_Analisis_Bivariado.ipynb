El riesgo cardiovascular es una preocupación de salud pública a nivel mundial. Herramientas como el Framingham Score han sido ampliamente utilizadas para estimar el riesgo de eventos cardiovasculares en diversas poblaciones. Sin embargo, su aplicabilidad en comunidades con características socioeconómicas y demográficas específicas, como el Distrito de Aguablanca en Cali, Colombia, requiere evaluación. ## Revisión Bibliográfica

A continuación, se presentan algunos artículos científicos que abordan la evaluación del riesgo cardiovascular y la aplicabilidad del Framingham Score en diferentes contextos:​

**1. Evaluación del Riesgo de Enfermedad Cardiovascular: Perspectivas desde Framingham**

Este artículo destaca que las funciones de riesgo de Framingham combinan información de factores como sexo, edad, presión arterial sistólica, colesterol total, colesterol HDL, tabaquismo y diabetes para estimar el riesgo de desarrollar enfermedades cardiovasculares en un período determinado, en los próximos 10 años.

**2. Estado Socioeconómico, Factores de Riesgo Cardíaco y Enfermedad Cardiovascular**

Este estudio revela que la hipertensión es el factor de riesgo más común en grupos de bajo nivel socioeconómico. Además, factores como antecedentes familiares de enfermedades cardiovasculares, tabaquismo e hiperlipidemia también son prevalentes en distintos grupos socioeconómicos.

**3. Asociación del Ingreso Familiar con el Riesgo de Enfermedad Cardiovascular en Adultos Estadounidenses**

Este artículo analiza cómo el bajo ingreso familiar, una medida del estado socioeconómico, puede estar asociado con un mayor riesgo de enfermedades cardiovasculares, destacando la importancia de considerar factores socioeconómicos en la evaluación del riesgo.

**4. Sobre la presión arterial y su relación para determinar riesgo caridovascular**

Este artículo pone en contexto por qué medir la presión diferencial puede ayudar a un equipo de atención médica a predecir el riesgo de padecer eventos cardíacos y vasculares, como ataques cardíacos o accidentes cerebrovasculares. Una presión diferencial superior a 60 se considera factor de riesgo para enfermedad cardíaca, especialmente en personas mayores.

# <a id="univariate-analysis"></a> 3.2 Análisis Univariado


Para este primer análisis, teniendo en cuenta la escala de Framingham y las referencias bibliográficas consultadas, fueron seleccionadas las siguientes variable claves: edad, tensión arterial sistólica, tensión arterial diastólica, peso y talla. Estas corresponden a las siguientes columnas del dataset:

* EDAD (Años cumplidos)
* 25_TenArtSis
* 26_TenArtDitlica
* 23_Peso
* 24_Talla

Para cada una de ellas:

1. Calcularemos estadísticas descriptivas (media, mediana, desviación estándar, valores mínimos y máximos).
2. Graficaremos su distribución con histogramas y boxplots para realizar las observaciones.


```python
# Seleccion de variables
features = ['EDAD (Años cumplidos)','23_Peso', '24_Talla', '25_TenArtSis','26_TenArtDitlica']

# Verificar registros con caracteres especiales dentro del formato numerico
df[features] = df[features].apply(pd.to_numeric, errors='coerce')
df_clean = df[features] .dropna(subset=features)
df_clean.describe().T
```

```python

# Generar gráficas para analizar la distribución
for col in df_clean.columns:
    if col in df_clean.columns:
        fig, axes = plt.subplots(1, 3, figsize=(14, 5))

        # Histograma con KDE (curva de densidad)
        sns.histplot(df_clean[col].dropna(), bins=30, kde=True, ax=axes[0], color="royalblue")
        axes[0].set_title(f"Distribución de {col}")
        axes[0].set_xlabel(col)
        axes[0].set_ylabel("Frecuencia")

        # Boxplot para detección de outliers
        sns.boxplot(x=df_clean[col], ax=axes[1], color="orange")
        axes[1].set_title(f"Boxplot de {col}")


        # Violinplot para visualizar la densidad de los datos
        sns.violinplot(x=df_clean[col], ax=axes[2], color="green")
        axes[2].set_title(f"Violinplot de {col}")

        plt.tight_layout()
        plt.show()
```

🔹 **Edad, peso y talla**

Podemos observar que la población analizada se encuentra entre los 9 y 115 años de edad (dato que podría llegar a ser erróneo) y que presentan tallas entre los 1.30m y 1.98m, siendo 1.58m la estatura media de la población analizada. En cuanto al peso, para una persona de 1.98 metros de altura, un peso saludable se encuentra en el rango de 67 a 89 kg según el Índice de Masa Corporal (IMC) normal (18.5-24.9). Teniendo esto en cuenta, podríamos afirmar que existen personas con sobrepeso en la población, dado que el valor máximo para esta variable es de 170 Kg.

🔹 **Tensión Arterial Sistólica y Diastólica (mmHg)**

Para realizar la interpretación de los valores relacionados con la presión arterial, podemos tener en cuenta la siguiente tabla, de la AHA - American Heart Association:

| Categoría                | Sistólica (mmHg) | Diastólica (mmHg) |
| ------------------------ | ---------------- | ----------------- |
| **Normal**               | Menos de 120     | Menos de 80       |
| **Elevada**              | 120-129          | Menos de 80       |
| **Hipertensión Etapa 1** | 130-139          | 80-89             |
| **Hipertensión Etapa 2** | 140 o más        | 90 o más          |
| **Crisis Hipertensiva**  | Más de 180       | Más de 120        |

Al observar los datos de la población en estudio, vemos que la media de la tensión (o presión) arterial sistólica tiene un valor de 126.1 mmHg, es decir, elevada. Además, se puede evidenciar un valor de p25 igual a 120 mmHg, o sea que sólo el 25% de la población tiene una presión sistólica normal (menor a 120). El resto de la población tiene un valor elevado e incluso en estado crítico llegando al valor máximo de 250 mmHg, representando una hipertención severa (o un posible error en los registros).

Del lado de la tensión arterial diastólica, podemos ver una media de 78 mmHg, lo que indica un valor elevado o normal, según la tabla de clasificación, que para un individuo particular, se debe evaluar en conjunto con la presión arterial sistólica y determinar si presenta hipertensión. El valor máximo de 152 mmHg indica un posible error en los datos o lo que podría ser un valor crítico de hipertensión.

# <a id="bivariate-analysis"></a> 3.3 Análisis Bivariado: características principales

Para el analisis bivariado, consideraremos en principio las métricas que de acuerdo a la literatura podrian explicar la clasificacion del riesgo cardiovascular: edad, peso, talla e IMC.

```python
# Boxplot de la variable 'EDAD (Años cumplidos)' por 'CLASIFICACION FINAL DEL RIESGO'
plt.figure(figsize=(10,6))
sns.boxplot(x='CLASIFICACION FINAL DEL RIESGO', y='EDAD (Años cumplidos)', data=df)
plt.title('Distribución de la EDAD por Clasificación de Riesgo Cardiovascular')
plt.xlabel('Clasificación Riesgo Cardiovascular')
plt.ylabel('EDAD (Años cumplidos)')
plt.show()
```

En términos generales, se observa un aumento de la mediana de edad a medida que el riesgo cardiovascular es mayor.
Pacientes con riesgo "MUY ALTO" tienden a ser mayores de 60 años.
Se observan outliers en edades tempranas (menores de 36 años) con clasificación de riesgo cardiovascular "MODERADO" a "MUY ALTO", lo que sugiere una medida de alerta en la población más joven.

La relación positiva entre edad y riesgo es esperable, puesto que el envejecimiento es un factor clave debido a la pérdida de elasticidad arterial, aumento de presión arterial y acumulación de otros factores de riesgo como hábitos.  En comunidades vulnerables como Aguablanca, el acceso limitado a salud preventiva puede llevar a diagnósticos tardíos en adultos mayores.


```python

# Boxplot de la variable 'Peso' por 'CLASIFICACION FINAL DEL RIESGO'
plt.figure(figsize=(10,6))
sns.boxplot(x='CLASIFICACION FINAL DEL RIESGO', y='23_Peso', data=df)
plt.title('Distribución del Peso por Clasificación de Riesgo Cardiovascular')
plt.xlabel('Clasificación Riesgo Cardiovascular')
plt.ylabel('Peso')
plt.show()
```

Se observa un rango amplio de pesos en todas las categorías de Clasificación de Riesgo Cardiovascular para esta muestra de la población, por tanto la variable Peso por sí sola es insuficiente para medir el riesgo cardiovascular.

Es importante destacar que se ha documentado en diversos estudios que la Obesidad y sobrepeso son factores de riesgo cardiovascular, puesto que aumentan la probabilidad de hipertensión, diabetes tipo 2 y enfermedad alterial.  Un peso alto combinado con tensión arterial elevada y resistencia a la insulina genera mayor probabilidad de eventos cardiovasculares.

En contexto con la población de Aguablanca, las dificultades económicas (pobreza a pobreza extrema) pueden llevar a una dieta deficiente, basada principalmente en carbohidratos y grasas saturadas.

Los Boxplot de talla por categoría de Riesgo no muestran diferencias significativas. Se puede observar que se presenta outliers en las categorías "MODERADO" Y "ALTO" para tallas por debajo de 137 cm, esto podría tratarse de menores de edad que posean uno o más factores de riesgo, sería importante hacer seguimiento a la población joven e infantil.

La nutricion infantil juega un papel importante en el desarrollo de enfermedades crónicas en la adultez.  En poblaciones vulnerables como Aguablanca puede presentarse una dieta poco saludable en la infancia, con poca fruta y verdura,lo que puede aumentar factores de riesgo en la adultez, también si los neonatos presentan bajo peso esto puede conllevar a una mayor presión arterial en adultos jóvenes.

Para aquellos pacientes que tienen peso "Bajo" a "Normal" y se encuentren clasificados con riesgo cardiovascular de "MODERADO" a "MUY ALTO", probablemente sean pacientes con otros factores de riesgo entre los cuales se debe examinar: hipertensión, diabetes, o malos hábitos como tabaquismo y/o alcoholismo.

## **Variables con Alta Correlación (Factores de riesgo claves)**

### **Perímetro Abdominal vs. Peso (0.58) y Perímetro de Cadera (0.82): Alta correlación positiva**

La **obesidad abdominal** es un mejor predictor de riesgo cardiovascular que el IMC.
Relacionado con diabetes tipo 2, hipertensión y dislipidemia.

### **Colesterol Total vs. LDL (0.78): Alta correlación positiva**
Es esperable, ya que el colesterol total se compone en gran parte de LDL.

Aumenta el riesgo de infarto y accidente cerebrovascular.

La TFG (Tasa de Filtración Glomerular) mide la función renal.

### **Hemoglobina Glicosilada vs. Glicemia (0.66)**
Relación positiva moderada, lo que indica que altos niveles de glucosa en sangre se reflejan en HbA1c elevada. Pacientes **diabéticos** tienen mayor riesgo cardiovascular debido a la inflamación crónica y daño endotelial.
HbA1c > 7% indica mal control glucémico.
[La hemoglobina glicosilada (HbA1c) es un análisis de sangre que mide el nivel promedio de glucosa en la sangre durante los últimos dos o tres meses. Se utiliza para diagnosticar la diabetes tipo 2 y la prediabetes]

### **Albuminuria vs. Cociente Albuminuria/Creatinuria (0.80): Relación esperada**
Si hay más albumina en orina, el cociente albuminuria/creatininuria también aumenta. La albuminuria es un **predictor de daño renal** y cardiovascular.
Signo temprano de nefropatía diabética e hipertensiva.

### **Tensión Arterial Sistólica vs. Tensión Arterial Diastólica (0.65)**
Relación esperado porque:
La presión sistólica es la presión que ejerce la sangre cuando el corazón late.
La presión diastólica es la presión que ejerce la sangre cuando el corazón está en reposo entre latidos.

Cuando la tensión sistólica sube, la diastólica también. Si la **presión arterial es alta se corre grave riesgo de complicaciones como infarto de corazón, accidente cerebrovascular, insuficiencia renal, otras enfermedades del corazón.**

Eliminamos las columnas que no aportan al modelo
Identificamos y manejamos valores nulos en las variables seleccionadas

```python
# Importar librerías necesarias
import pandas as pd
import numpy as np

# Variables seleccionadas para el modelo
variables_modelo = ["Risk_Level", "28_HemoGlicosilada", "84_Cociente albuminuria/creatinuria", 
                    "Perimetro abdominal", "35_TFG", "EDAD (Años cumplidos)", 
                    "25_TenArtSis", "26_TenArtDitlica"]

# Filtrar solo las variables relevantes
df_modelo = df_clean[variables_modelo]

# Verificar valores nulos
print("📌 Cantidad de valores nulos por columna:")
print(df_modelo.isnull().sum())

# Eliminar filas donde Risk_Level esté nulo (porque es la variable objetivo)
df_modelo = df_modelo.dropna(subset=["Risk_Level"])
```

## **B. Impotación de Valores Faltantes**

Donde hay valores faltantes en variables numéricas los rellenamos con estrategias estadísticas como imputar con la mediana.

```python
from sklearn.impute import SimpleImputer

# Crear un imputador con la mediana
imputer = SimpleImputer(strategy="median")

# Aplicar la imputación solo a las variables numéricas
df_modelo.iloc[:, 1:] = imputer.fit_transform(df_modelo.iloc[:, 1:])
```

```python
# Mostrar algunas filas antes de la imputación
df_modelo_before = df_clean[variables_modelo].copy()
df_modelo_after = df_modelo.copy()

# Comparar valores antes y después
print("🔍 Comparación de valores antes y después de la imputación:")
for col in variables_modelo[1:]:  # Omitimos Risk_Level porque no se imputa
    missing_before = df_modelo_before[col].isnull().sum()
    missing_after = df_modelo_after[col].isnull().sum()
    
    if missing_before > 0:
        print(f"📌 {col}: {missing_before} nulos antes → {missing_after} nulos después")
        print(f"Ejemplo de valores después de imputación:\n {df_modelo_after[col].head(10)}\n")
```
